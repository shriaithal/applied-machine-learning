#### K-Fold for Model Evaluation

```python
#User defined functon for K-Fold cross validatoin
def build_confusion_matrix(classifier):
    
    k_fold = KFold(n_splits=5, random_state=None, shuffle=False)
    scores = []
    confusion = np.array([[0,0],[0,0]])

    for train_ind, test_ind in k_fold.split(train_news):
        train_text = train_news.iloc[train_ind]['headline_text'] 
        train_y = train_news.iloc[train_ind]['encoded_label']
    
        test_text = train_news.iloc[test_ind]['headline_text']
        test_y = train_news.iloc[test_ind]['encoded_label']
        
        classifier.fit(train_text,train_y)
        predictions = classifier.predict(test_text)
        
        confusion += confusion_matrix(test_y,predictions)
        score = f1_score(test_y,predictions)
        scores.append(score)
    
    print('Total statements classified:{}'.format(len(train_news)))
    print('Score:{}'.format(sum(scores)/len(scores)))
    print('Score length:{}'.format(len(scores)))
    cm = confusion
    sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g')
    plt.tight_layout()
    plt.title('Confusion matrix', y=1.1)
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
```

#### Feature Engineering: Political Affiliation Vector Approach 2

To label documents based on political affiliation the next approach used is Doc2Vec.

**Doc2Vec**: Doc2vec is an unsupervised algorithm to generate vectors for sentence/paragraphs/documents. Doc2Vec is an adaptation of word2vec model. The vectors generated by doc2vec can be used for tasks like finding similarity between sentences/paragraphs/document

**Doc2Vec Parameters**:
1. Using Distributed bag of words dm=0
2. Feature Dimension - 500
3. Set min_count=2 ignoring words in text whose frequence is lower than 2

**Steps Followed**:
1. Identify different categories in political affiliation dictionary and group the words.
2. Calculate the political affiliation polarity of the headline text by find the frequency of occuurance.
3. Assign political affiliation label based on the highest polarity.
4. Tag each document based on political polarity of the document.
5. Run Doc2Vec model
6. Run SVM to perform multi-class classification

**Classification Report**
SVM classification yields only 39%accuracy.

```python
#Group dictionary based on political affiliation factor
pos_df = political_dict[political_dict['Positive'] != 0]
pos_df.drop(['Negative', 'Strong', 'Weak', 'Active', 'Passive'], axis=1, inplace=True)

neg_df = political_dict[political_dict['Negative'] != 0]
neg_df.drop(['Positive', 'Strong', 'Weak', 'Active', 'Passive'], axis=1, inplace=True)

strong_df = political_dict[political_dict['Strong'] != 0]
strong_df.drop(['Positive', 'Negative', 'Weak', 'Active', 'Passive'], axis=1, inplace=True)

weak_df = political_dict[political_dict['Weak'] != 0]
weak_df.drop(['Positive', 'Negative', 'Strong', 'Active', 'Passive'], axis=1, inplace=True)

active_df = political_dict[political_dict['Active'] != 0]
active_df.drop(['Positive', 'Negative', 'Strong', 'Weak', 'Passive'], axis=1, inplace=True)

passive_df = political_dict[political_dict['Passive'] != 0]
passive_df.drop(['Positive', 'Negative', 'Strong', 'Weak', 'Active'], axis=1, inplace=True)
```

```python
#nltk.download('popular')
#Lemmetize the words in headline_text
def normalize_word(word):
    return nltk.WordNetLemmatizer().lemmatize(word).lower()
```

```python
#Calculate the polarity of the headline. For each word in the headline, find the frequency in the dictionary. Sum the score and normalize by max value
def get_polatiry_for_text(text, affiliation_dict, max_val):
        tokens = [normalize_word(word) for word in nltk.word_tokenize(text)]
        #print(affiliation_dict)
        #print(sum([token in affiliation_dict for token in tokens]))
        return sum([token in affiliation_dict for token in tokens]) % max_val
```

```python
#Calculate text polarity whether headline text is positive, negative, strong, weak, active or passive
def get_political_affiliation_features(data):
    political_affiliation = []
    for index, row in data.iterrows():
        headline_polarity = []
        headline_polarity.append(get_polatiry_for_text(row['headline_text'], np.array(pos_df['Word']), max(pos_df['Positive'])))
        headline_polarity.append(get_polatiry_for_text(row['headline_text'], np.array(neg_df['Word']), max(neg_df['Negative'])))
        headline_polarity.append(get_polatiry_for_text(row['headline_text'], np.array(strong_df['Word']), max(strong_df['Strong'])))
        headline_polarity.append(get_polatiry_for_text(row['headline_text'], np.array(weak_df['Word']), max(weak_df['Weak'])))
        headline_polarity.append(get_polatiry_for_text(row['headline_text'], np.array(active_df['Word']), max(active_df['Active'])))
        headline_polarity.append(get_polatiry_for_text(row['headline_text'], np.array(passive_df['Word']), max(passive_df['Passive'])))
        political_affiliation.append(headline_polarity)
    return political_affiliation
```

```python
#Once the polarity is found, find the maximum value and label the text based on max score
label_dict = {0:'Positive', 1:'Negative', 2:'Strong', 3:'Weak', 4:'Active', 5:'Passive'}

def get_political_affiliation_labels(data):
    political_affiliation_labels = []
    for i in range(len(data)):
        max_index = np.argmax(data[i])
        political_affiliation_labels.append(label_dict[max_index])
    return political_affiliation_labels
```

```python
#tag each headline text with the label found in previous step
def tag_headline(data):
    labelled_headlines = []
    for index, row in data.iterrows():
        labelled_headlines.append(TaggedDocument(words=word_tokenize(row['headline_text']), tags=[row['political_affiliation_label']]))
    return labelled_headlines
```

```python
#append labels to train set
political_affiliation_train = get_political_affiliation_features(df)
df['political_affiliation_label'] = get_political_affiliation_labels(political_affiliation_train)
#df.head(5)
```

```python
#tag train data and run doc2vec model
labelled_headlines_train = tag_headline(df)
#print(labelled_headlines_train[:3])

doc2vec_model_train = Doc2Vec(documents = labelled_headlines_train, dm=0, num_features=500, min_count=2, size=15, window=4)
```

```python
#Method to convert doc2vec model and tagged documents into vectors training and testing
def create_vector_for_learning(model, tagged_docs):
    #documents = tagged_docs.values
    targets, regressors = zip(*[(doc.tags[0], model.infer_vector(doc.words, steps=20)) for doc in tagged_docs])
    return targets, regressors
```

**Word2Vec**: Doc2vec is an unsupervised algorithm to generate vectors for sentence/paragraphs/documents. Doc2Vec is an adaptation of word2vec model. The vectors generated by doc2vec can be used for tasks like finding similarity between sentences/paragraphs/document

**Word2Vec Parameters**:
1. num_features = 6      : Word vector dimensionality                      
2. min_word_count = 40   : Minimum word count                        
3. num_workers = 4       : Number of threads to run in parallel
4. context = 10          : Context window size                                                                                    
5. downsampling = 1e-3   : Downsample setting for frequent words

**Steps Followed To Create Political Affiliation Vector**:
1. Split each headline into array of words
2. Find the intersection of words in headline and political dictionary
3. For the intersection set, calculate the emotions by summing.
4. Normalize this vector of emotions
5. Append to cleaned data set.

**Steps to create word2vec model**
1. Tokenize headline text to words using regex to split words based on space
2. Create a Word2Vec model based from the Word Tokens
3. Derive vectors from the Word2Vec model for each headline. The vector dimension is 6
4. Find the dot prodcut between polarity matrix and the vector generated from word2vec model
5. The dot product signifies the similarity score between the political affiliaition dictionary and the headline text

```python
# method to split article to list of words
def string_to_word_list(text):
    #print(text)
    words = re.split(r'\W+', text)
    return words
```

```python
# method to find intersection of two list
def intersect(x, y):
    return list(set(x) & set(y))
```

```python
# normalize vector value
def normalize_list(score_list):
    max_value = max(score_list)
    min_value = min(score_list)
    for i in range(0, len(score_list)):
        if (max_value - min_value) != 0:
            score_list[i] = '%.3f'%((score_list[i] - min_value) / (max_value - min_value))
        if (max_value - min_value) == 0:
            score_list[i] = '%.3f'%0
    return score_list
```

```python
# method to form political affiliation vector
def calculate_political_affliation_vector(intersection_list):
   # affliation_score = []
    pos = 0
    neg = 0
    active = 0
    passive = 0
    strong = 0
    weak = 0

    for i in range(len(intersection_list)):
        row = political_dict[political_dict['Word'] == intersection_list[i]]
        #print(row)
        pos = pos + (float(row['Positive']))
        neg = neg + (float(row['Negative']))
        strong = strong + (float(row['Strong']))
        weak = weak + (float(row['Weak']))
        active = active + (float(row['Active']))
        passive = passive + (float(row['Passive']))

    affliation_score = [pos, neg, strong, weak, active, passive]
    return affliation_score
```

```python
# get political affiliation for every article in the dataset
def get_political_affliation(articles):
    dictionary = political_dict['Word']
    political_affliation_vector = []
    for i in range(len(articles)):
        #print (articles[i])
        words = string_to_word_list(articles[i])
        intersection_list = intersect(words, dictionary)
        affliation_score = calculate_political_affliation_vector(intersection_list)
        #print(affliation_score)
        affliation_score = normalize_list(affliation_score)
        political_affliation_vector.append(affliation_score)
    return political_affliation_vector
```

```python
num_features = 6      # Word vector dimensionality                      
min_word_count = 40   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel
context = 10          # Context window size                                                                                    
downsampling = 1e-3   # Downsample setting for frequent words

#Create Word2Vec Model. Get Vector from the model
model = word2vec.Word2Vec(df['word_tokens'], workers=num_workers, size=num_features, min_count = min_word_count, window = context, sample = downsampling)
vectors = model.wv
```

```python
#Get feature vector from the trained wor2vec model for each headline_text
def get_w2v_features(w2v_model, words):

    index2word_set = set(w2v_model.wv.vocab.keys())  # words known to model
    
    featureVec = np.zeros(w2v_model.vector_size, dtype="float32")
    
    # Initialize a counter for number of words in a review
    nwords = 0
    # Loop over each word in the comment and, if it is in the model's vocabulary, add its feature vector to the total
    for word in words:
        if word in index2word_set: 
            featureVec = np.add(featureVec, w2v_model[word])
            nwords += 1.

    # Divide the result by the number of words to get the average
    if nwords > 0:
        featureVec = np.divide(featureVec, nwords)
    return featureVec
```

```python
#For each headline text, generate a Word2Vec Model
w2v = list(map(lambda sen_group:
                                      get_w2v_features(model, sen_group),
                                      df['word_tokens']))
```

```python
w2v_norm = []
def normalize_w2v(w2v):
    for i in range(len(w2v)):
        word_vec = w2v[i]
        norm = np.linalg.norm(word_vec)
        if norm == 0: 
           w2v_norm.append(word_vec)
        if norm != 0:
            w2v_norm.append(word_vec/norm)
```

#### Feature Engineering: Political Affiliation Vector Approach 4

In this method, cosine similarity between dictionary of words and TF-IDF vector is calculated.

**TFIDF Vector**: Tf-idf stands for term frequency-inverse document frequency. This weight is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus. 

**Cosine Similarity**: The cosine similarity between two vectors (or two documents on the Vector Space) is a measure that calculates the cosine of the angle between them. This metric is a measurement of orientation and not magnitude, it can be seen as a comparison between documents on a normalized space because we’re not taking into the consideration only the magnitude of each word count (tf-idf) of each document, but the angle between the documents.

**Steps Followed**
1. Create a corpus of the political dictionary
2. Create TF-IDF Vector of the dictionary
3. Create TF-IDF Vector of every headline text
4. Find cosine similarity between TF-IDF Dictionary Vector and Headline Text
